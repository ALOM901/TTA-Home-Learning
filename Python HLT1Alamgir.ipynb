Task 1
Write a program that does the following:
a) Stores a random number (1-10) in a variable – see hint below.
b) Asks a user for their name and stores this in a variable.
c) Asks a user to guess the number between 1 and 10.
d) Tells the user whether they have guessed correctly.

Solution
1. import random library
2. Ask python to pick a random number between 1-10
3. Ask user what is their name
4. Ask user to guess a number
5. Create an IF statement to determine if the number the user guessed is the same as the random number generated by python 

```python
import random    
rand_num =  random.randint(1,10)
user_name  = input("what is your name")
guess_number = int(input("Select a number between 1 and 10"))

if rand_num == guess_number:
    print("Well done you guessed correct")

else: 

    print(user_name,"You guessed wrong, it was not:", guess_number, "it was", rand_num)
```

Solution
1. Store the value of the bike against a variable
2. Create a while loop which subtracts 10% of the bikes value untill the value reaches £1000

```python
bike_cost = 2000
year_start = 0

while bike_cost >= 1000:
    print("The bike is worth", bike_cost, "in year", year_start)
    bike_cost *=0.90
    year_start += 1
    
```

Task 5
Write a program which will ask for two numbers from a user. Then offer a menu to the user giving them a choice of operator:e.g. – Enter “a” if you want to add“b” if you want to subtract.
Include +, -, /, *, ** square (to the power of). Once the user has selected which operator they wish to use, perform the calculation.

Solution
1. Ask the user to pick a number
2. Ask the user to pick a second number
3. Ask the user to pick an operator
4. Create an IF statement to carry out the math operation using the users choice of numbers and choice of operator

```python

num_1 = int(input("Please enter a number: "))
num_2 = int(input("Please enter a second number: "))
sign = input("Please enter a operator choice(+,-,/,*,**,%,//):")

if sign == "+":  
    print(num_1 + num_2)

elif sign == "-":
    print(num_1 - num_2)

elif sign == "/":
    print(num_1 / num_2)

elif sign == "*":
    print(num_1 * num_2)

elif sign == "**":
    print(num_1 ** num_2)

elif sign == "%":
    print(num_1 % num_2)
elif sign == "//":
    print(num_1 // num_2)
else:
    print("An appropriate operator was not selected!")
```
